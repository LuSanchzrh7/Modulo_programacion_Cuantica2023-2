```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from scipy import constants as cs
from IPython.display import HTML
```

```python
x = np.linspace(0, 1, 1000)
t = np.linspace(0, 50000, 200)
n=[1,2,3,4]
```

```python
def funcion_de_onda(n,x, t,A=np.sqrt(30),m=cs.m_e,a=1):
    p_1=-A*np.sqrt(2)
    p_2=(((-1)**n-1)/(np.pi**3*n**2))**2
    p_3=np.exp((-pow(n,2)*np.pi*cs.hbar*t*1j)/(2*m*a**2))
    p_4=np.sqrt(2/a)*np.sin((n*np.pi*x)/a)
    return p_1*p_2*p_3*p_4
```

```python
X, T = np.meshgrid(x, t)
```

```python
plt.imshow(funcion_de_onda(1, X, T).real, cmap='viridis')
plt.colorbar()
```
